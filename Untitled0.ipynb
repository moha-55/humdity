```python
import pandas as pd
import numpy as np
import time
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix, classification_report
```

```python
# Load dataset
data_url = "https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data"
column_names = [
    "age", "workclass", "fnlwgt", "education", "education-num", "marital-status",
    "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss",
    "hours-per-week", "native-country", "income"
]
data = pd.read_csv(data_url, names=column_names, na_values=" ?", skipinitialspace=True)
```

```python
data
```

```python
# Drop rows with missing values
data.dropna(inplace=True)
```

```python
data
```

```python
# Keep only 2 classes: <=50K and >50K (already binary)
# Sample 1500 rows total
data = data.sample(n=1500, random_state=42)
```

```python
data.describe()
```

```python
# Encode categorical features
label_encoders = {}
for column in data.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    data[column] = le.fit_transform(data[column])
    label_encoders[column] = le
```

```python
# Split features and labels
X = data.drop("income", axis=1)
y = data["income"]
```

```python
# Normalize features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.25, random_state=42)
```

```python
# --- Classifier 1: SVM ---
print("\nTraining SVM...")
start_train = time.time()
svm_clf = SVC()
svm_clf.fit(X_train, y_train)
end_train = time.time()

start_test = time.time()
y_pred_svm = svm_clf.predict(X_test)
end_test = time.time()

print("SVM Training Time: {:.4f} seconds".format(end_train - start_train))
print("SVM Testing Time: {:.4f} seconds".format(end_test - start_test))
print("SVM Confusion Matrix:\n", confusion_matrix(y_test, y_pred_svm))
```

```python
# --- Classifier 2: Logistic Regression ---
print("\nTraining Logistic Regression...")
start_train = time.time()
logreg_clf = LogisticRegression(max_iter=1000)
logreg_clf.fit(X_train, y_train)
end_train = time.time()

start_test = time.time()
y_pred_logreg = logreg_clf.predict(X_test)
end_test = time.time()

print("LogReg Training Time: {:.4f} seconds".format(end_train - start_train))
print("LogReg Testing Time: {:.4f} seconds".format(end_test - start_test))
print("LogReg Confusion Matrix:\n", confusion_matrix(y_test, y_pred_logreg))
```
